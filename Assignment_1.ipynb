```python
import numpy as np
import matplotlib.pyplot as plt

#training data
trxdata = np.loadtxt('/content/hw1xtr.dat')
trydata = np.loadtxt('/content/hw1ytr.dat')

#test data
texdata = np.loadtxt('/content/hw1xte.dat')
teydata = np.loadtxt('/content/hw1yte.dat')


#Plot training data
plt.figure(figsize=(8, 4))
plt.plot(trxdata, trydata, 'bx', label='Train')
plt.xlabel('Input Feature')
plt.ylabel('Desired Output')
plt.title('Training Data')
plt.legend()
plt.show()

#Plot test data
plt.figure(figsize=(8, 4))
plt.plot(texdata, teydata, 'gx', label='Test')
plt.xlabel('Input Feature')
plt.ylabel('Desired Output')
plt.title('Test Data')
plt.legend()
plt.show()
```

```python


#Adding column vector of 1's to the features
trxdata = trxdata.reshape(-1, 1)
column_vector_1 = np.ones((trxdata.shape[0], 1))
X_new = np.concatenate((trxdata, column_vector_1), axis=1)
#print(X_new)

# weight vector, w = (X^T X)^-1 X^T y
X_T = np.transpose(X_new)
w = np.linalg.inv(X_T @ X_new) @ X_T @ trydata

# Plot the training data
plt.plot(trxdata, trydata, 'bx', label='Training Data')

y_pred = X_new @ w

# Plot linear regression line
plt.plot(trxdata, y_pred, 'r-', label='Linear Regression Line')
plt.xlabel('Features (X)')
plt.ylabel('Outputs (Y)')
plt.title('Training Data vs Linear Regression')
plt.legend()

plt.show()

#average error calculation
m = X_new.shape[0]
total_error = 0
for i in range(m):

    y_new_i = np.dot(X_new[i], w)  # y= w^⊤ * x_i
    error_i = (y_new_i - y_pred[i]) ** 2
    total_error += error_i
avg_error = total_error / m
print(f'average error for training set is {avg_error}')
```
